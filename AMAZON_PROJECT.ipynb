### COMBINING DATA FROM 15 - 25

```python
df_all.head()
```

## 1. Standardize order_date

```python
df_all['order_date'] = pd.to_datetime(df_all['order_date'], errors='coerce')
df_all['order_date'] = df_all['order_date'].dt.strftime('%Y-%m-%d')
```

## 2. Clean original_price_inr

```python
# Remove '₹' symbol and commas, replace 'Price on Request' with NaN
df_all['original_price_inr'] = (df_all['original_price_inr']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .replace('Price on Request', np.nan))

# Convert the column to numeric, coercing errors (non-convertible to NaN)
df_all['original_price_inr'] = pd.to_numeric(df_all['original_price_inr'], errors='coerce')
```

## 3. Customer ratings standardization

```python
df_all['customer_rating'] = df_all['customer_rating'].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
df_all['customer_rating'] = df_all['customer_rating'].fillna(df_all['customer_rating'].median())
```

## 4. Standardize customer_city

```python
city_map = {
    'bangalore': 'Bengaluru', 'bengaluru': 'Bengaluru',
    'mumbai': 'Mumbai', 'bombay': 'Mumbai',
    'delhi': 'Delhi', 'new delhi': 'Delhi'
}
df_all['customer_city'] = df_all['customer_city'].str.lower().replace(city_map)
df_all['customer_city'] = df_all['customer_city'].str.title()
```

## 5. Boolean columns standardization

```python
true_vals = ['true','yes','y','1']
false_vals = ['false','no','n','0']
for col in ['is_prime_member','is_prime_eligible','is_festival_sale']:
    df_all[col] = df_all[col].astype(str).str.lower()
    df_all[col] = df_all[col].apply(lambda x: True if x in true_vals else False)
```

## 6. Product category standardization

```python
cat_map = {
    'electronics':'Electronics', 'electronic':'Electronics', 'electronics & accessories':'Electronics'
}
df_all['category'] = df_all['category'].str.lower().replace(cat_map)
df_all['category'] = df_all['category'].str.title()
```

## 7. Clean delivery_days

```python
df_all['delivery_days'] = df_all['delivery_days'].replace({'Same Day':0, 'same day':0})
df_all['delivery_days'] = df_all['delivery_days'].astype(str).str.extract(r'(\d+)')[0].astype(float)
df_all.loc[df_all['delivery_days'] < 0, 'delivery_days'] = np.nan
df_all.loc[df_all['delivery_days'] > 30, 'delivery_days'] = np.nan
df_all['delivery_days'] = df_all['delivery_days'].fillna(df_all['delivery_days'].median())
```

## 8. Handle duplicate transactions

```python
dupe_mask = df_all.duplicated(subset=['customer_id','product_id','order_date','original_price_inr'], keep=False)
df_all['is_duplicate'] = dupe_mask
```

## 9. Correct outlier prices

```python
max_price = df_all['original_price_inr'].quantile(0.99)
df_all.loc[df_all['original_price_inr'] > max_price*10, 'original_price_inr'] = df_all['original_price_inr']/100
```

## 10. Standardize payment methods

```python
pay_map = {
    'upi' : 'UPI', 'phonepe':'UPI', 'googlepay':'UPI',
    'credit card':'Credit Card', 'credit_card':'Credit Card', 'cc':'Credit Card',
    'cash on delivery':'Cash on Delivery', 'cod':'Cash on Delivery', 'c.o.d':'Cash on Delivery'
}
df_all['payment_method'] = df_all['payment_method'].str.lower().replace(pay_map)
df_all['payment_method'] = df_all['payment_method'].str.title()
```

# SQL DATABASE

```python
import sqlite3
import pandas as pd
```

```python
# Connect to SQLite database
conn = sqlite3.connect('amazon_india.db')
cursor = conn.cursor()

# Step 1: Create tables
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
    customer_id INTEGER PRIMARY KEY,
    customer_city TEXT,
    customer_state TEXT,
    customer_tier TEXT,
    customer_age_group TEXT,
    customer_spending_tier TEXT,
    is_prime_member BOOLEAN
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS products (
    product_id INTEGER PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    subcategory TEXT,
    brand TEXT,
    product_weight_kg REAL,
    product_rating REAL
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS time_dimension (
    date TEXT PRIMARY KEY,
    day INTEGER,
    month INTEGER,
    quarter INTEGER,
    year INTEGER,
    day_of_week INTEGER,
    is_weekend BOOLEAN
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS transactions (
    transaction_id INTEGER PRIMARY KEY,
    order_date TEXT,
    customer_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    original_price_inr REAL,
    discount_percent REAL,
    discounted_price_inr REAL,
    subtotal_inr REAL,
    delivery_charges REAL,
    final_amount_inr REAL,
    customer_city TEXT,
    customer_state TEXT,
    customer_tier TEXT,
    customer_spending_tier TEXT,
    customer_age_group TEXT,
    payment_method TEXT,
    delivery_days INTEGER,
    delivery_type TEXT,
    is_prime_member BOOLEAN,
    is_festival_sale BOOLEAN,
    customer_rating REAL,
    return_status TEXT,
    order_month TEXT,
    order_year INTEGER,
    order_quarter INTEGER,
    product_weight_kg REAL,
    product_rating REAL,
    is_prime_eligible BOOLEAN,
    is_duplicate BOOLEAN
);
""")


conn.commit()

# Step 2: Prepare your dataframes from df_all

df_customers = df_all[['customer_id', 'customer_city', 'customer_state', 'customer_tier',
                       'customer_age_group', 'customer_spending_tier', 'is_prime_member']].drop_duplicates()

df_products = df_all[['product_id', 'product_name', 'category', 'subcategory', 'brand',
                      'product_weight_kg', 'product_rating']].drop_duplicates()

df_time_dimension = df_all[['order_date']].drop_duplicates()
df_time_dimension.rename(columns={'order_date': 'date'}, inplace=True)
df_time_dimension['date'] = pd.to_datetime(df_time_dimension['date'], errors='coerce')
df_time_dimension = df_time_dimension.dropna(subset=['date'])
df_time_dimension['day'] = df_time_dimension['date'].dt.day
df_time_dimension['month'] = df_time_dimension['date'].dt.month
df_time_dimension['quarter'] = df_time_dimension['date'].dt.quarter
df_time_dimension['year'] = df_time_dimension['date'].dt.year
df_time_dimension['day_of_week'] = df_time_dimension['date'].dt.dayofweek
df_time_dimension['is_weekend'] = df_time_dimension['day_of_week'] >= 5
df_time_dimension['date'] = df_time_dimension['date'].dt.strftime('%Y-%m-%d')

df_transactions = df_all[['transaction_id', 'order_date', 'customer_id', 'product_id', 'quantity',
                          'original_price_inr', 'discount_percent', 'discounted_price_inr', 'subtotal_inr',
                          'delivery_charges', 'final_amount_inr', 'customer_city', 'customer_state',
                          'customer_tier', 'customer_spending_tier', 'customer_age_group', 'payment_method',
                          'delivery_days', 'delivery_type', 'is_prime_member', 'is_festival_sale',
                          'customer_rating', 'return_status', 'order_month', 'order_year', 'order_quarter',
                          'product_weight_kg', 'product_rating', 'is_prime_eligible', 'is_duplicate']].drop_duplicates()

df_transactions['order_date'] = pd.to_datetime(df_transactions['order_date'], errors='coerce')
df_transactions = df_transactions.dropna(subset=['order_date'])
df_transactions['order_date'] = df_transactions['order_date'].dt.strftime('%Y-%m-%d')

# Convert Period columns to string for SQLite compatibility
df_transactions['order_month'] = df_transactions['order_month'].astype(str)

# Step 3: Load dataframes to SQLite
df_customers.to_sql('customers', conn, if_exists='replace', index=False)
df_products.to_sql('products', conn, if_exists='replace', index=False)
df_time_dimension.to_sql('time_dimension', conn, if_exists='replace', index=False)
df_transactions.to_sql('transactions', conn, if_exists='replace', index=False)

# Step 4: Create indexes after loading
cursor.execute("CREATE INDEX IF NOT EXISTS idx_order_date ON transactions(order_date);")
cursor.execute("CREATE INDEX IF NOT EXISTS idx_customer_id ON transactions(customer_id);")
cursor.execute("CREATE INDEX IF NOT EXISTS idx_product_id ON transactions(product_id);")

conn.commit()
conn.close()
```
